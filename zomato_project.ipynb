```python
def parse_dates(train):
    train['order_time'] = pd.to_datetime(train['order_time'], format = "%Y-%m-%d %X")
    train['order_date'] = pd.to_datetime(train['order_date'], format = "%Y-%m-%d %X")
    train['allot_time'] = pd.to_datetime(train['allot_time'], format = "%Y-%m-%d %X")
    train['accept_time'] = pd.to_datetime(train['accept_time'], format = "%Y-%m-%d %X")

    return train
```

```python
def feat_eng(data):
  data['accept_order_diff'] = (data['accept_time'] - data['allot_time']).dt.total_seconds()
  data['total_dist'] = data.first_mile_distance + data.last_mile_distance
  data.reassigned_order = data['reassigned_order'].fillna(value = 0)
  data.reassignment_method = data.reassignment_method.fillna(value = 'none')
  data.reassignment_reason = data.reassignment_reason.fillna(value = 'none')
  data.lifetime_order_count = data.lifetime_order_count.fillna(0)
  return data
```

```python
train = feat_eng(train)
```

```python
for i in range(len(train)):
  if pd.isna(train['alloted_orders'][i]) and pd.isna(train['delivered_orders'][i]) and pd.isna(train['undelivered_orders'][i]):
    train['alloted_orders'][i]=train['delivered_orders'][i]=train['undelivered_orders'][i]=0
  elif pd.isna(train['delivered_orders'][i]) and pd.isna(train['undelivered_orders'][i]):
    train['delivered_orders'][i] = 0
    train['undelivered_orders'][i] = train['alloted_orders'][i]
```

```python
def preprocessing(data):
  data.session_time = data.session_time.fillna(data.session_time.median())
  return data
```

```python
train = preprocessing(train)
```

```python
from imblearn.under_sampling import RandomUnderSampler
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import StratifiedKFold
```

```python
y = train.cancelled
X = train.drop('cancelled', axis = 1)
```

```python
numeric_cols = [
'first_mile_distance',
'last_mile_distance',
'lifetime_order_count',
'alloted_orders',
'delivered_orders',
'session_time','total_dist','accept_order_diff','allot_delay'
]
```

```python
rs = RobustScaler()
X[numeric_cols] = rs.fit_transform(X[numeric_cols])
```

```python
other_cols=['first_mile_distance_count','first_mile_distance_mean','first_mile_distance_max','first_mile_distance_min','first_mile_distance_sum',
               'last_mile_distance_mean','last_mile_distance_max','last_mile_distance_min','last_mile_distance_sum',
               'lifetime_order_count_mean','lifetime_order_count_max','lifetime_order_count_min','lifetime_order_count_sum',
               'alloted_orders_mean','alloted_orders_max','alloted_orders_min','alloted_orders_sum',
               'delivered_orders_mean','delivered_orders_max','delivered_orders_min','delivered_orders_sum',
               'session_time_mean','session_time_max','session_time_min','session_time_sum',
               'total_dist_mean','total_dist_max','total_dist_min','total_dist_sum',
               'accept_order_diff_mean','accept_order_diff_max','accept_order_diff_min','accept_order_diff_sum',
               'allot_delay_mean','allot_delay_max','allot_delay_min','allot_delay_sum'
              ]
```

```python
X[other_cols] = rs.fit_transform(X[other_cols])
```

```python
kfold = StratifiedKFold(n_splits = 10, shuffle = True, random_state = 41)
```

```python
sampler = RandomUnderSampler(sampling_strategy=0.5,random_state=101)
X_sampled,y_sampled=sampler.fit_resample(X,y)
```

```python
import lightgbm as lgb
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score,accuracy_score,classification_report
```

```python
test_preds_xgb = []

for fold, (trn_ind, val_ind) in enumerate(kfold.split(X_sampled, y_sampled)):
    print(f"=====================fold: {fold + 1}=====================")

    X_train, y_train = X.iloc[trn_ind], y.iloc[trn_ind]
    X_valid, y_valid = X.iloc[val_ind], y.iloc[val_ind]

    model = XGBClassifier(
        n_estimators = 800, eval_metric = 'auc', tree_method="hist",
        enable_categorical = True, use_label_encoder = False
    )

    model.fit(X_train, y_train, early_stopping_rounds = 150, eval_set = [(X_valid, y_valid)], verbose = 100)

    preds_proba = model.predict_proba(X_valid)
    pred = model.predict(X_valid)
    roc = roc_auc_score(y_valid, preds_proba[:,1])
    auc = accuracy_score(y_valid,pred)
    print(F'fold {fold + 1}: ROC AUC {roc}')
    print(F'fold{fold +1}: Accuracy {auc}')
```
